# Stemmer

```python
from nltk.tokenize import word_tokenize
from nltk.stem import LancasterStemmer, SnowballStemmer, PorterStemmer, WordNetLemmatizer
```

```python
word_list = ['writing', 'playing', 'played', 'riding', 'drives', 'breaking', 'lives', 'running', 'given', 'chosen', 'drawing']
```

```python
lancaster = LancasterStemmer() # cepat, akurasi kurang
snowball = SnowballStemmer('english') # bisa untuk banyak bahasa
porter = PorterStemmer() # akurat, lambat
```

```python
for word in word_list:
    print(f"word = {word}")
    print(f"Lancaster = {lancaster.stem(word)}")
    print(f"Snowball = {snowball.stem(word)}")
    print(f"Porter = {porter.stem(word)}")
    print("================")
```

# Lemmatizer

```python
lemmatizer = WordNetLemmatizer()
word = 'better'

# pos = 'a' adjective, 'v' verb, 'n' noun, 'r' adverb
verb = lemmatizer.lemmatize(word, 'v')
adv = lemmatizer.lemmatize(word, 'r')
noun = lemmatizer.lemmatize(word, 'n')
adj =lemmatizer.lemmatize(word, 'a')

print(verb, adv, noun, adj)
```

```python
sentence = 'Quickly kicking quirky kangaroos caused chaos.'
word_list1 = word_tokenize(sentence)
word_list1
```

```python
lowered_case_list = [word.lower() for word in word_list1 if word.isalpha()]
lowered_case_list
```

```python
word_list = ['good']
# synsets = wordnet.synsets('good.n.01')
# synsets.definition()
# synsets.lemmas()[0].name()
# synsets.lemmas()[0].antonyms()

# synsets = wordnet.synsets(word_list)
for word in lowered_case_list:
    synonyms = []
    antonyms = []
    synsets = wordnet.synsets(word)
    for syn in synsets:
        for lemma in syn.lemmas():
            synonyms.append(lemma.name())
            
            for ant in lemma.antonyms():
                antonyms.append(ant.name())
    print(f"word = {word}")
    print(f"synonyms = {synonyms}")
    print(f"antonyms = {antonyms}")
    print("=====================")
```
