```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
glove_dict = {}

with open("glove.6B.100d.txt", "r", encoding="utf-8") as file:
    for line in file:
        values = line.split(" ")
        word = values[0]
        vectors = np.array(values[1:], dtype="float32")
        glove_dict[word] = vectors
        
```

```python
# glove_dict["king"]
if "king" in glove_dict.keys():
    print(glove_dict["king"])
```

```python
keys = list(glove_dict.keys())
values = np.array(list(glove_dict.values()))
values
```

```python
vector = glove_dict["king"] - glove_dict["man"] + glove_dict["woman"]
vector
```

```python
similarity = cosine_similarity([vector], values).flatten()
similarity[0]
```

```python
# ascending
top_idx = similarity.argsort()[-3:][::-1]
top_idx[0]
```

```python
for idx in top_idx:
    print(f"keys: {keys[idx]}")
    print(f"score: {similarity[idx]}")
    print()
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
```

```python
df = pd.read_csv("mtsamples.csv")
df.head()
```

```python
docs = df["transcription"].dropna().tolist()
docs
```

```python
def get_tokens(sentence):
    return [token.lower() for token in word_tokenize(sentence)]
```

```python
tokenized_docs = []
for sentence in docs:
    tokens = get_tokens(sentence)
    tokenized_docs.append(tokens)
```

```python
skipgram = Word2Vec(
    tokenized_docs,
    min_count=1, 
    window=5, 
    vector_size=100,
    sg=1
    )
cbow = Word2Vec(
    tokenized_docs,
    min_count=1, 
    window=5, 
    vector_size=100
)
```

```python
skipgram.wv["heart"]
```

```python
skipgram.wv.similarity("heart", "obesity")
```

```python
skipgram.wv.most_similar("heart", topn=5)
```

```python
cbow.wv.similarity("heart", "obesity")
```

```python
cbow.wv.most_similar("heart", topn=5)
```
