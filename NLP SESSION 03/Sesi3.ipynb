```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tag import pos_tag
import string
```

```python
sentence = "Elon Musk, the billionaire polymath who has become as synonymous with orbital mechanics as he is with internet-breaking social media posts, is currently navigating a 2026 that finds him at the helm of a newly forged $1.25 trillion empire following the historic merger of SpaceX and xAI, a move that aims to solve artificial intelligence’s terrestrial energy crisis by launching massive solar-powered data centers into orbit while he simultaneously pivots Tesla’s focus away from luxury sedans like the Model S and X toward a future dominated by millions of Optimus humanoid robots, all while he edges closer to becoming the world's first trillionaire as a high-stakes SpaceX IPO looms on the horizon, scheduled—with his characteristic flair for the dramatic—to coincide with a rare planetary alignment and his fifty-fifth birthday this June."

words_list = word_tokenize(sentence)
eng_stopwords = stopwords.words('english')
removed_stopwords = [token for token in words_list if token is not eng_stopwords]
# removed_stopwords

punctuation = string.punctuation
removed_punctuation = [token for token in removed_stopwords if token not in punctuation]
removed_punctuation
```

```python
tag = pos_tag(removed_punctuation)
tag
```

```python
from nltk.chunk import ne_chunk
```

```python

chunk_list = ne_chunk(tag)


```

# Frequency Distribution

```python
from nltk.probability import FreqDist
```

```python
# fdist = FreqDist(sentence)
fdist = FreqDist(words_list)
for token, count in fdist.most_common(5):
    print(f"{token}: {count}")
```
